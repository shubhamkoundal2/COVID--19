The dataset you provided contains COVID-19 data for different countries/regions. It includes information such as the number of confirmed cases, deaths, recoveries, active cases, and various percentages related to these numbers. The dataset also includes data from the previous week and categorizes each country/region into the respective WHO Region. This dataset can be used for analyzing and comparing the COVID-19 situation across different countries/regions, identifying trends, and assessing the impact of the pandemic. With the help of Python and data visualization libraries, the dataset can be further explored and visualized to gain insights into the distribution and trends of COVID-19 cases globally.

## **1.Which country/region has the highest number of confirmed cases?**

```python
# Sort the DataFrame by the 'Confirmed' column in descending order and select the top 10 rows
top_10_confirmed = df.sort_values('Confirmed', ascending=False).head(10)

# Create a bar chart to visualize the top 10 countries/regions with the highest number of confirmed cases
plt.figure(figsize=(10, 6))
plt.bar(top_10_confirmed['Country/Region'], top_10_confirmed['Confirmed'])
plt.xlabel('Country/Region')
plt.ylabel('Number of Confirmed Cases')
plt.title('Top 10 Countries/Regions with Highest Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **2.How many new deaths were reported worldwide?**

```python
new_deaths = df['New deaths']

# Calculate the total new deaths worldwide
total_new_deaths = new_deaths.sum()

# Print the total new deaths
print("Total new deaths worldwide:", total_new_deaths)

# Create a line plot to visualize the new deaths over time
plt.figure(figsize=(10, 6))
plt.plot(new_deaths)
plt.title('New Deaths Worldwide')
plt.xlabel('Days')
plt.ylabel('Number of New Deaths')
plt.show()
```

## **3.Which WHO Region has the highest number of new deaths?**

```python
# Group the data by WHO Region and calculate the sum of new deaths
region_deaths = df.groupby('WHO Region')['New deaths'].sum()

# Find the region with the highest number of new deaths
max_deaths_region = region_deaths.idxmax()

# Create a pie chart
plt.figure(figsize=(15, 6))
plt.pie(region_deaths, labels=region_deaths.index, autopct='%1.1f%%')
plt.title('New Deaths by WHO Region')
# Display the chart
plt.show()
```

## **4.Which country/region has the highest recovery rate of top 10?**

```python
# Sort the DataFrame by recovery rate in descending order
df = df.sort_values(by='Recovery Rate', ascending=False)

# Select the top 10 countries/regions with the highest recovery rates
top_10 = df.head(10)

# Visualize the recovery rates using a bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_10['Country/Region'], top_10['Recovery Rate'])
plt.xlabel('Country/Region')
plt.ylabel('Recovery Rate (%)')
plt.title('Top 10 Countries/Regions with the Highest Recovery Rates')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## **5.In which WHO region are most of the countries/regions located?**

```python
# Count the number of countries/regions in each WHO region
region_counts = df['WHO Region'].value_counts()

# Plotting the results
plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar')
plt.xlabel('WHO Region')
plt.ylabel('Number of Countries/Regions')
plt.title('Number of Countries/Regions in Each WHO Region')
plt.show()
```
